## Organização de dados em variáveis e estruturas lógicas

```python
# Aí entra o conceito mais importante da programação: a Variável. Por que a gente não usa simplesmente os números soltos nos 
# cálculos? Porque no mundo real, a informação muda o tempo todo! Se o faturamento da empresa mudar amanhã, eu não quero ter 
# que caçar esse número em 50 linhas de código e mudar um por um na mão. Usando a variável, eu mudo o valor em um único lugar 
# e, automaticamente, todos os cálculos que usam aquela 'etiqueta' são atualizados. É inteligência e ganho de tempo.

# Comentar: Selecione e pressione Ctrl+K seguido de Ctrl+C.
# Descomentar: Selecione e pressione Ctrl+K seguido de Ctrl+U.

faturamento = 1000
custo = 400
Lucro = faturamento - custo

print("faturamento:", faturamento)
print("custo:", custo)
print("Lucro:", Lucro)

# Mas atenção para um detalhe importante: o Python é muito organizado. Se eu tentar fazer um cálculo usando o 'Faturamento' 
# antes de dizer para o Python quanto ele vale, o programa vai travar. Primeiro a gente define a variável, dá um valor pra 
# ela, e só depois a gente usa. É a ordem natural das coisas.
faturamento = 1000
custo = 400
Lucro = faturamento - custo

print("faturamento:", faturamento)
print("custo:", custo)
print("Lucro:", Lucro)
```

```python

faturamento = 1100
custo = 400
lucro = faturamento - custo

novas_vendas = 100

faturamento = faturamento + novas_vendas

print('Faturamento:', faturamento)
print('Custo:', custo)
print('Lucro:', lucro)
```

```python
# Tipos de variáveis:

# int = numeros inteiros
# float = numeros com casas decimais
# strings = textos
# Boolean = booleanos (True, False)


faturamento = 1100
custo = 400
lucro = faturamento - custo

novas_vendas = 100

faturamento = faturamento + novas_vendas
print('Lucro:', lucro)

# Pq usa "." em vez da virgula, pois é um padrão americano.
imposto = 0.15 * faturamento # float
print('imposto:', imposto)

lucro = faturamento - custo - imposto

print('Faturamento:', faturamento)
print('Custo:', custo)
print('Lucro:', lucro)

# Vamos supor q preciso mandar uma mensagem pra alguem com os resultados
mensagem = "O faturamento da loja foi de 1200"

teve_lucro = True # boolean

# Pra gente fechar esse conceito, vamos falar de Boas Práticas. Na Hashtag, a gente preza muito pelo código limpo. Evitem 
# nomes de variáveis como x, y ou a. Usem nomes intuitivos! Se é faturamento, escreva faturamento. Se é imposto, escreva 
# imposto. Isso ajuda você e qualquer outra pessoa que for ler seu código depois.
```
