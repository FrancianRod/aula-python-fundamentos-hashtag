Pessoal, o projeto de hoje é um Sistema de Performance B2B. A finalidade dele é automatizar a política de bônus da empresa. Em vez de calcularmos na mão, vamos criar um código que recebe os dados de venda, analisa se o vendedor bateu a meta e decide automaticamente qual bônus ele vai receber. É o Python agindo como um consultor estratégico dentro da operação.

```python
# --- [TÓPICO 3: VARIÁVEIS E MEMÓRIA] ---
# Definindo as "gavetas" de memória com as regras fixas de negócio.
meta_ouro = 50000.00
meta_prata = 30000.00
taxa_ouro = 0.15   # Representa 15%
taxa_prata = 0.10  # Representa 10%
taxa_bronze = 0.05 # Representa 5% (Incentivo)

# valor = 60000
# bonus_ouro = valor + (valor*taxa_ouro)
# print(bonus_ouro)
# valor_int = str(bonus_ouro)
# print(valor_int)
# tipo = type(valor_int)
# print(tipo)

# print("Valor feito com o acrecimo do bonus foi", bonus_ouro)

# Tipos de Variaveis
# int = numeros interios ----- 10
# float = numero cm casas decimais ---- 10.1
# string = texto ------- "tipo de variavel texto"
# boolean = True, False 
```

```python
# Estruturação de scripts para leitura e processamento
# --- [TÓPICO 4: INTERAÇÃO COM USUÁRIO] ---
# Capturando dados dinâmicos e aplicando o conceito de Casting (Conversão).
# input é uma ferramenta do python q permite vc pegar informações do usuário
# A Natureza do Input: "Tudo o que o usuário digita no teclado, o Python entende como String (texto). Mesmo que ele digite '5000', para o Python 
# aquilo é apenas um desenho de caracteres.
nome_vendedor = input("Digite o nome do consultor: ")
# F-Strings: Esse f antes das aspas no print. Nos permite 'injetar' variáveis diretamente dentro do texto de um jeito muito elegante e rápido.
venda_texto = input(f"Informe o total vendido por {nome_vendedor}: R$ ")

# IMPORTANTE: Convertendo o texto do input para número decimal (float)
valor_vendas = float(venda_texto)
```

```python
# --- [TÓPICO 5: O CÉREBRO - LÓGICA DE DECISÃO IF/ELIF/ELSE] ---
# Construção das regras de decisão para classificação automática.

# Estrutura de Controle: O if é uma bifurcação. O código só entra ali se a condição for True (Verdadeira). Se for False, ele pula para a próxima.
# A Importância do elif: Poderíamos usar vários if seguidos, mas o elif (else if) é mais eficiente. Ele só é testado se o if de cima falhar. 
# Isso economiza processamento e evita que o vendedor caia em duas categorias ao mesmo tempo.
# Indentação (O mais importante): Reparem no espaço (TAB) antes do categoria =. No Python, esse espaço define o que pertence ao if. 
# É o que chamamos de bloco de código. Sem isso, o Python não sabe onde a regra termina.

if valor_vendas >= meta_ouro:
    categoria = "OURO"
    bonus = valor_vendas * taxa_ouro
    mensagem = "Desempenho excepcional! Superou a meta máxima."
elif valor_vendas >= meta_prata:
    categoria = "PRATA"
    bonus = valor_vendas * taxa_prata
    mensagem = "Bom desempenho! Meta batida com sucesso."
else:
    categoria = "BRONZE"
    bonus = valor_vendas * taxa_bronze
    mensagem = "Abaixo da meta esperada. Foco no próximo mês!"

# O script agora está agindo como um gestor: ele analisa o dado, compara com a regra e entrega uma decisão (o bônus) e um feedback (a mensagem).

print(f"O vendedor {nome_vendedor} ficou na categoria {categoria} e seu desempenho foi {mensagem}")
```

```python
# --- [TÓPICO 7: CONCLUSÃO E DESAFIO] ---
# Aplicação final da lógica e entrega do relatório de tomada de decisão.
percentual_atingido = (valor_vendas / meta_ouro) * 100

print("\n" + "="*40)
print(f"RELATÓRIO DE PERFORMANCE: {nome_vendedor}")
print(f"CATEGORIA ALCANÇADA: {categoria}")
print(f"PORCENTAGEM DA META OURO: {percentual_atingido:.2f}%")
print(f"BÔNUS CALCULADO: R$ {bonus:,.2f}")
print(f"FEEDBACK AUTOMÁTICO: {mensagem}")
print("="*40)
# print("\nDESAFIO: Como você adicionaria uma categoria DIAMANTE para vendas acima de R$ 100k?")
```

```python
# --- [TÓPICO 6: SIMULAÇÃO DE ERROS E DEBUGGING (PARTE A)] ---
# Introduzimos o 'try' aqui para proteger todo o processo de entrada e cálculo.

# Um sistema profissional não pode 'travar' e mostrar um erro feio para o usuário se ele digitar algo errado. O try é o nosso 'escudo'.
# Exceção (ValueError): Eu digo ao Python: Tente fazer isso. Se der um erro de valor (ValueError), não trave! Em vez disso, execute esse aviso amigável.
# Isso é o que separa um script de estudante de uma ferramenta corporativa robusta.

try:
    
    # --- [TÓPICO 4: INTERAÇÃO COM USUÁRIO] ---
    # Capturando dados dinâmicos e aplicando o conceito de Casting (Conversão).
    nome_vendedor = input("Digite o nome do consultor: ")
    venda_texto = input(f"Informe o total vendido por {nome_vendedor}: R$ ")
    
    # IMPORTANTE: Convertendo o texto do input para número decimal (float)
    valor_vendas = float(venda_texto)

    # --- [TÓPICO 5: O CÉREBRO - LÓGICA IF/ELIF/ELSE] ---
    # Construção das regras de decisão para classificação automática.
    if valor_vendas >= meta_ouro:
        categoria = "OURO"
        bonus = valor_vendas * taxa_ouro
        mensagem = "Desempenho excepcional! Superou a meta máxima."
    elif valor_vendas >= meta_prata:
        categoria = "PRATA"
        bonus = valor_vendas * taxa_prata
        mensagem = "Bom desempenho! Meta batida com sucesso."
    else:
        categoria = "BRONZE"
        bonus = valor_vendas * taxa_bronze
        mensagem = "Abaixo da meta esperada. Foco no próximo mês!"

    # --- [TÓPICO 7: CONCLUSÃO E DESAFIO] ---
    # Aplicação final da lógica e entrega do relatório de tomada de decisão.
    percentual_atingido = (valor_vendas / meta_ouro) * 100

    print("\n" + "="*40)
    print(f"RELATÓRIO DE PERFORMANCE: {nome_vendedor}")
    print(f"CATEGORIA ALCANÇADA: {categoria}")
    # Formatação de Moeda: Usar :.2f serve para garantir que o bônus sempre mostre duas casas decimais, essencial para relatórios financeiros profissionais.
    print(f"PORCENTAGEM DA META OURO: {percentual_atingido:.2f}%")
    print(f"BÔNUS CALCULADO: R$ {bonus:,.2f}")
    print(f"FEEDBACK AUTOMÁTICO: {mensagem}")
    print("="*40)
    # print("\nDESAFIO: Como você adicionaria uma categoria DIAMANTE para vendas acima de R$ 100k?")

# --- [TÓPICO 6: SIMULAÇÃO DE ERROS E DEBUGGING (PARTE B)] ---
# Tratamento de erro caso o aluno/usuário digite letras onde o Python espera números.
except ValueError:
    print("\n[ERRO DE SISTEMA]: Entrada inválida detectada.")
    print("Por favor, insira apenas números para o valor de vendas (ex: 5000.50).")
```
