Sistema de análise de performance - Hashtag

```python
# Metas e Taxas de Bônus referente ao sistema de performance

meta_ouro = 50000.00
meta_prata = 30000.00
taxa_ouro = 0.15 # Representa 15% de Bônus
taxa_prata = 0.10 # Representa 10% de Bônus
taxa_bronze = 0.05 # Representa 5% de Bônus

# valor_venda = 60000
# bonus_ouro = valor_venda * taxa_ouro
# print(bonus_ouro)

# print("Valor de bonus recebido pelo vendedor foi ", bonus_ouro)


# 4 tipos de variáveis
# int = nuemros inteiros
# float = numeros com casas decimais
# strings = texto -> "" or ''
# Boolean = True, False
```

```python
# Interação do usuários, dados de vendas e vendedor

nome_vendedor = input("Digite o nome do vendedor: ")
venda_texto = input(f"Informe o total vendido por {nome_vendedor}: R$ ")

# Conversão do meu valor recebido de venda, str -> float
valor_venda = float(venda_texto)
```

```python
# Lógica de Decisão, if/elif/else

if valor_venda >= meta_ouro:
    categoria = "Ouro"
    bonus = valor_venda * taxa_ouro
    mensagem = "Desempenho excepcional!"
elif valor_venda >= meta_prata:
    categoria = "Prata"
    bonus = valor_venda * taxa_prata
    mensagem = "Bom desempenho!"
else:
    categoria = "Bronze"
    bonus = valor_venda * taxa_bronze
    mensagem = "Abaixo da meta esperada!"

print(f"O vendedor {nome_vendedor} ficou na categoria {categoria} e seu desempenho foi {mensagem}")
```

```python
# Aplicação final da lógica e entrega do relatório de tomada de decisão

percentual_atingido = (valor_venda / meta_ouro) * 100

print("\n" + "=" * 50)
print(f"RELATÓRIO DE PERFORMANCE: {nome_vendedor}")
print(f"CATEGORIA ALCANÇADA: {categoria}")
print(f"PORCENTAGEM DA META OURO: {percentual_atingido:.2f}")
print(f"BÔNUS CALCULADO: R$ {bonus:.2f}")
print(f"FEEDBACK AUTOMÁTICO: {mensagem}")
print("=" * 50)
```

```python
# SIMULAÇÃO DE ERROS E DEBUGGING

try:

    # Interação do usuários, dados de vendas e vendedor

    nome_vendedor = input("Digite o nome do vendedor: ")
    venda_texto = input(f"Informe o total vendido por {nome_vendedor}: R$ ")

    # Conversão do meu valor recebido de venda, str -> float
    valor_venda = float(venda_texto)

    # Lógica de Decisão, if/elif/else

    if valor_venda >= meta_ouro:
        categoria = "Ouro"
        bonus = valor_venda * taxa_ouro
        mensagem = "Desempenho excepcional!"
    elif valor_venda >= meta_prata:
        categoria = "Prata"
        bonus = valor_venda * taxa_prata
        mensagem = "Bom desempenho!"
    else:
        categoria = "Bronze"
        bonus = valor_venda * taxa_bronze
        mensagem = "Abaixo da meta esperada!"

    print(f"O vendedor {nome_vendedor} ficou na categoria {categoria} e seu desempenho foi {mensagem}")

    # Aplicação final da lógica e entrega do relatório de tomada de decisão

    percentual_atingido = (valor_venda / meta_ouro) * 100

    print(f"RELATÓRIO DE PERFORMANCE: {nome_vendedor}")
    print(f"CATEGORIA ALCANÇADA: {categoria}")
    print(f"PORCENTAGEM DA META OURO: {percentual_atingido:.2f}")
    print(f"BÔNUS CALCULADO: R$ {bonus:.2f}")
    print(f"FEEDBACK AUTOMÁTICO: {mensagem}")

except ValueError:
    print("\n[Erro de sistema]: Entrada inválida detectada.")
    print("Por favor, insira apenas números para o valor de vendas (ex: 6000.55)")
```
